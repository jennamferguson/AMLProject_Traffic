```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('traffic_weather.csv')
```

```python
data.isnull().sum()
```

```python
# correlation matrix
numeric_data = data.select_dtypes(include=['number'])
correlation_matrix = numeric_data.corr()
correlation_matrix
```

```python


plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap="coolwarm", vmin=-1, vmax=1)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
#scatter plot of tavg and volume
sns.scatterplot(x='Speed', y='Volume', data=data)
plt.show()
```

```python
data.describe()
```

```python
data.dtypes
```

```python
hourly_volume = data.groupby('Hour')['Volume'].mean()
plt.figure(figsize=(10, 6))
hourly_volume.plot(kind='line', marker='o')
plt.title('Average Traffic Volume by Hour')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Traffic Volume')
plt.grid()
plt.show()
```

```python
# 2. Day of Week Patterns
day_of_week_volume = data.groupby('Day of Week')['Volume'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=day_of_week_volume.index, y=day_of_week_volume.values, palette="viridis")
plt.title('Average Traffic Volume by Day of the Week')
plt.xlabel('Day of the Week (0=Monday, 6=Sunday)')
plt.ylabel('Average Traffic Volume')
plt.show()
```

```python
# 3. Holiday vs Non-Holiday
holiday_volume = data.groupby('Holiday')['Volume'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=['Non-Holiday', 'Holiday'], y=holiday_volume.values, palette="coolwarm")
plt.title('Average Traffic Volume: Holiday vs Non-Holiday')
plt.ylabel('Average Traffic Volume')
plt.show()
```

```python
# 4. Monthly Patterns
monthly_volume = data.groupby('Month')['Volume'].mean()
plt.figure(figsize=(10, 6))
monthly_volume.plot(kind='bar', color='skyblue')
plt.title('Average Traffic Volume by Month')
plt.xlabel('Month')
plt.ylabel('Average Traffic Volume')
plt.grid()
plt.show()
```

```python
# 5. Directional Traffic
directional_volume = data.groupby('Direction')['Volume'].mean()
plt.figure(figsize=(10, 6))
sns.barplot(x=directional_volume.index, y=directional_volume.values, palette="Set2")
plt.title('Average Traffic Volume by Direction')
plt.ylabel('Average Traffic Volume')
plt.xlabel('Direction')
plt.show()
```

## Weather EDA


```python
# 1. Correlation Analysis
plt.figure(figsize=(10, 8))
correlation_columns = ['Volume', 'tavg', 'tmin', 'tmax', 'prcp', 'snow', 'wspd']
correlation_matrix = data[correlation_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Between Traffic Volume and Weather Variables')
plt.show()
```

```python
# 2. Precipitation/Snow Analysis
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['prcp'] > 0, y=data['Volume'], palette="Blues", notch=True)
plt.xticks([0, 1], ['No Precipitation', 'Precipitation'])
plt.title('Traffic Volume Distribution: Precipitation vs. No Precipitation')
plt.ylabel('Traffic Volume')
plt.xlabel('Precipitation')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=data['snow'] > 0, y=data['Volume'], palette="Purples", notch=True)
plt.xticks([0, 1], ['No Snow', 'Snow'])
plt.title('Traffic Volume Distribution: Snow vs. No Snow')
plt.ylabel('Traffic Volume')
plt.xlabel('Snow')
plt.show()
```

```python
# 3. Temperature Effects
plt.figure(figsize=(10, 6))
sns.scatterplot(x='tavg', y='Volume', data=data, alpha=0.5)
plt.title('Traffic Volume vs. Average Temperature')
plt.xlabel('Average Temperature (°C)')
plt.ylabel('Traffic Volume')
plt.grid()
plt.show()
```

```python
# 4. Wind Speed Impact
plt.figure(figsize=(10, 6))
sns.scatterplot(x='wspd', y='Volume', data=data, alpha=0.5, color='orange')
plt.title('Traffic Volume vs. Wind Speed')
plt.xlabel('Wind Speed (km/h)')
plt.ylabel('Traffic Volume')
plt.grid()
plt.show()
```

```python
# Traffic Volume by Road, Direction, and Hour of the Day

# Group the data by Road Name, Direction, and Hour and calculate the average traffic volume
road_direction_hour_volume = data.groupby(['Intersection Name', 'Direction', 'Hour'])['Volume'].mean().reset_index()

# Create separate plots for each road
roads = data['Intersection Name'].unique()
for road in roads:
    road_data = road_direction_hour_volume[road_direction_hour_volume['Intersection Name'] == road]
    plt.figure(figsize=(12, 6))
    sns.lineplot(data=road_data, x='Hour', y='Volume', hue='Direction', marker='o')
    plt.title(f'Hourly Traffic Volume for {road}')
    plt.xlabel('Hour of the Day')
    plt.ylabel('Average Traffic Volume')
    plt.legend(title='Direction')
    plt.grid()
    plt.show()
```
